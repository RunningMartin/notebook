## 颜色条

`colorbar`函数可以创建简单的颜色条。

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,10,1000)
i=np.sin(x)*np.cos(x[:,np.newaxis])
plt.imshow(i)
plt.colorbar()
```

### 配置颜色条

- 配置颜色方案：`plt`可以通过`cmap`选择配色方案。

```python
plt.imshow(i,cmap='gray')
```

`plt.cm`中可以找到所有可用的配色方案。关于配色方案选择讨论可以查看：
- https://journals.plos.org/ploscompbiol/article?id=10.1371/journal.pcbi.1003833
- https://matplotlib.org/1.4.1/users/colormaps.html

常见的配色方案有三种：
- 顺序配色方案：由一组连续的颜色构成，如`binary`或`viridis`。
- 互逆配色方案：由两种互补的颜色构成，表示正反，如`RdBu`或`PuOr`。
- 定性配色方案：随机选择的一组颜色，如`jet`或`rainbow`。

推荐使用`cubehelix`、`viridis`或`RdBu`。

- 离散型颜色条：颜色条也支持离散型颜色条，便于显示离散数据，其方法是使用`plt.cm.get_cmap()`函数，传入相应的配色方案和区间数量即可。

```python
plt.imshow(i,cmap=plt.cm.get_cmap('Blues',6))
plt.colorbar()
plt.clim(-1,1)
```

```python
plt.imshow(i,cmap=plt.cm.get_cmap('Blues',10))
plt.colorbar()
plt.clim(-1,1)
```

- 调整颜色条刻度：`plt.clim`配置颜色刻度表范围。

```python
plt.imshow(i,cmap=plt.cm.get_cmap('Blues',10))
plt.colorbar()
plt.clim(-2,2)
```

### 案例：手写数字

```python
# 加载手写体数字
from sklearn.datasets import load_digits
digits = load_digits(n_class=6)
# figsize调试图片尺寸
fig,ax = plt.subplots(8,8,figsize=(6,6))
# ax.flat中存储了8*8个元素，用于显示手写体
for i ,axi in enumerate(ax.flat):
    axi.imshow(digits.images[i],cmap='binary')
    axi.set(xticks=[],yticks=[])
```
