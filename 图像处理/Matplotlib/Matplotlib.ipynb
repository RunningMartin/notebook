## 简易线形图

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
# 切换风格
plt.style.use('seaborn-whitegrid')
# figure 是一个容器
fig = plt.figure()
ax = plt.axes()
```

```python
import numpy as np
x = np.linspace(0,10,1000)
fig = plt.figure()
ax = plt.axes()
ax.plot(x,np.sin(x))
```

### 风格调整

- 颜色

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,np.sin(x-1),color='blue')# 颜色名称
ax.plot(x,np.sin(x-2),color='g')# 颜色缩写 rgbcmyk
ax.plot(x,np.sin(x-3),color='0.75')# 灰度值 0-1
ax.plot(x,np.sin(x-4),color='#FFDD44')# 十六进制 RRGGBB
ax.plot(x,np.sin(x-5),color='chartreuse')# HTML颜色名称
ax.plot(x,np.sin(x-6),color=(1.0,0.2,0.3))# RGB元组，0-1
```

- 线条：线条简写可参考[官方文档](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)。

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,x,linestyle='solid')# 实线 简写-
ax.plot(x,x+1,linestyle='dashed')# 虚线 简写 --
ax.plot(x,x+2,linestyle='dashdot')# 点虚线，简写-.
ax.plot(x,x+3,linestyle='dotted')# 实点线 简写：
ax.plot(x,x+4,linestyle='-')# 实线 简写-
ax.plot(x,x+5,linestyle='--')# 虚线 简写 --
ax.plot(x,x+6,linestyle='-.')# 点虚线，简写-.
ax.plot(x,x+7,linestyle=':')# 实点线 简写：
```

- 编码简写：采用`RGB`与`CMYK`简写形式。

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,x,'-g')
ax.plot(x,x+1,'--c')
ax.plot(x,x+2,'-.k')
ax.plot(x,x+3,':r')
```

### 坐标轴调整

`plt.xlim`和`plt.ylim`用于调整x轴和y轴的上下限。

```python
fig = plt.figure()
ax = plt.axes()
# 调整上下限
plt.xlim(-1,6)
plt.ylim(0,5)
ax.plot(x,x,'-g')
```

`plt.xlim`和`plt.ylim`也支持传入逆序的信息，用于逆序显示坐标轴刻度值。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
plt.xlim(6,-1)
plt.ylim(5,0)
ax.plot(x,x,'-g')
```

设置x、y轴的界限还可以通过`plt.axis([xmin, xmax, ymin, ymax])`函数。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
plt.axis([0,5,-2,2])
ax.plot(x,np.sin(x),'-g')
```

`plt.axis()`还支持传入`tight`、`equal`参数，调整`x`轴和`y`轴大小。
- `tight`：图形收紧坐标，不留空白。
- `equal`:x轴和y轴单位长度相同。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.plot(x,np.sin(x),'-g')
plt.axis('tight')
```

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.plot(x,np.sin(x),'-g')
plt.axis('equal')
```

### 设置图像标签

- 图形标题：`plt.title()`

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.plot(x,np.sin(x),'-g')
ax.set_title('sin(x)')
plt.axis('tight')
```

- 坐标轴标签：`plt.xlabel`、`plt.ylabel`。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.set_title('sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.plot(x,np.sin(x),'-g')
plt.axis('tight')
```

- 简易图例：使用`plot`函数时，需要用`label`属性标记。然后用`plt.legend()`即可生成图例。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.set_title('sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.plot(x,np.sin(x),'-g',label='sin(x)')
ax.plot(x,np.cos(x),':r',label='cos(x)')
plt.axis('tight')
plt.legend()
```

`ax.set()`支持一次性设置所有的属性。

```python
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2),xlabel='x',ylabel='sin(x)',title='label sin(x)')
```

## 简单散点图

- `plt.plot`画图。

```python
x = np.linspace(0, 10, 30)
ax = plt.axes()
ax.plot(x,np.sin(x), 'o', color='black')
```

```python
x = np.linspace(0, 10, 30)
ax = plt.axes()
ax.plot(x,np.sin(x), '-p', color='green',
        markersize=15, 
        linewidth=4,
        markerfacecolor='white',
        markeredgecolor='gray',
        markeredgewidth=2)
```

- `plt.scatter`：画散点图，`scatter`与`plot`相比，具有更高的灵活性，可以控制每个散点的属性。但是由于`scatter`需要对每个数据都进行渲染，因此针对大型数据，`plot`性能更好。

```python
plt.scatter(x, np.sin(x), marker='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
# c是每个点的颜色，s为尺寸，alpha表示透明度，便于显示重叠部分
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,cmap='viridis')
# 显示颜色条
plt.colorbar(); 
```
