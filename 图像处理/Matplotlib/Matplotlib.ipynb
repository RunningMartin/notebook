## 简易线形图

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
# 切换风格
plt.style.use('seaborn-whitegrid')
# figure 是一个容器
fig = plt.figure()
ax = plt.axes()
```

```python
import numpy as np
x = np.linspace(0,10,1000)
fig = plt.figure()
ax = plt.axes()
ax.plot(x,np.sin(x))
```

### 风格调整

- 颜色

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,np.sin(x-1),color='blue')# 颜色名称
ax.plot(x,np.sin(x-2),color='g')# 颜色缩写 rgbcmyk
ax.plot(x,np.sin(x-3),color='0.75')# 灰度值 0-1
ax.plot(x,np.sin(x-4),color='#FFDD44')# 十六进制 RRGGBB
ax.plot(x,np.sin(x-5),color='chartreuse')# HTML颜色名称
ax.plot(x,np.sin(x-6),color=(1.0,0.2,0.3))# RGB元组，0-1
```

- 线条：线条简写可参考[官方文档](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)。

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,x,linestyle='solid')# 实线 简写-
ax.plot(x,x+1,linestyle='dashed')# 虚线 简写 --
ax.plot(x,x+2,linestyle='dashdot')# 点虚线，简写-.
ax.plot(x,x+3,linestyle='dotted')# 实点线 简写：
ax.plot(x,x+4,linestyle='-')# 实线 简写-
ax.plot(x,x+5,linestyle='--')# 虚线 简写 --
ax.plot(x,x+6,linestyle='-.')# 点虚线，简写-.
ax.plot(x,x+7,linestyle=':')# 实点线 简写：
```

- 编码简写：采用`RGB`与`CMYK`简写形式。

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,x,'-g')
ax.plot(x,x+1,'--c')
ax.plot(x,x+2,'-.k')
ax.plot(x,x+3,':r')
```

### 坐标轴调整

`plt.xlim`和`plt.ylim`用于调整x轴和y轴的上下限。

```python
fig = plt.figure()
ax = plt.axes()
# 调整上下限
plt.xlim(-1,6)
plt.ylim(0,5)
ax.plot(x,x,'-g')
```

`plt.xlim`和`plt.ylim`也支持传入逆序的信息，用于逆序显示坐标轴刻度值。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
plt.xlim(6,-1)
plt.ylim(5,0)
ax.plot(x,x,'-g')
```

设置x、y轴的界限还可以通过`plt.axis([xmin, xmax, ymin, ymax])`函数。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
plt.axis([0,5,-2,2])
ax.plot(x,np.sin(x),'-g')
```

`plt.axis()`还支持传入`tight`、`equal`参数，调整`x`轴和`y`轴大小。
- `tight`：图形收紧坐标，不留空白。
- `equal`:x轴和y轴单位长度相同。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.plot(x,np.sin(x),'-g')
plt.axis('tight')
```

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.plot(x,np.sin(x),'-g')
plt.axis('equal')
```

### 设置图像标签

- 图形标题：`plt.title()`

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.plot(x,np.sin(x),'-g')
ax.set_title('sin(x)')
plt.axis('tight')
```

- 坐标轴标签：`plt.xlabel`、`plt.ylabel`。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.set_title('sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.plot(x,np.sin(x),'-g')
plt.axis('tight')
```

- 简易图例：使用`plot`函数时，需要用`label`属性标记。然后用`plt.legend()`即可生成图例。

```python
fig = plt.figure()
ax = plt.axes()
# 逆序显示
ax.set_title('sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.plot(x,np.sin(x),'-g',label='sin(x)')
ax.plot(x,np.cos(x),':r',label='cos(x)')
plt.axis('tight')
plt.legend()
```

`ax.set()`支持一次性设置所有的属性。

```python
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2),xlabel='x',ylabel='sin(x)',title='label sin(x)')
```

## 误差线

- 基本误差线：`plt.errorbar(x,y,err,fmt='')`。`fmt`的数据同format。

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,50)
dy = 0.8
y=np.sin(x) + dy* np.random.randn(50)
fig = plt.figure()
ax = plt.axes()
ax.errorbar(x,y,yerr=dy,fmt='.k')
```

```python
ax = plt.axes()
# color指定数据点颜色 ecolor指定误差的颜色，elinewidth指定误差线颜色
ax.errorbar(x,y,yerr=dy,fmt='o',color='black',ecolor='lightgray',elinewidth=3,capsize=0)
```

- 连续误差:`使用fill_between`和`plot`实现

```python
from sklearn.gaussian_process import GaussianProcessRegressor
# 定义模型和要画的数据
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)
# 计算高斯过程拟合结果
gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, y_std = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit = 2 * np.sqrt(y_std)  # 置信区间 2*sigma ~ 95% 
```

```python
fig,ax= plt.subplots(3)
# 画点
ax[0].plot(xdata,ydata,'or')
ax[1].plot(xdata,ydata,'or')
ax[2].plot(xdata,ydata,'or')
# 画拟合线
ax[1].plot(xfit,yfit,'-',color='gray')
ax[2].plot(xfit,yfit,'-',color='gray')
# 画连续误差 x轴坐标，y轴正负误差，填充即可
ax[2].fill_between(xfit, yfit - dyfit, yfit + dyfit,color='gray', alpha=0.2)
ax[0].set_xlim(0,10)
ax[1].set_xlim(0,10)
ax[2].set_xlim(0,10)
```

## 密度图与等高线图

- 等高线图：`plt.contour`，配色方案可以在`plt.cm`中查找。

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
x= np.linspace(0,5,50)
y= np.linspace(0,4,50)
# 生成x，y轴的网格数据
x,y = np.meshgrid(x,y)
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
z=f(x,y)
```

```python
ax = plt.axes()
# 默认采用实线表示正数，虚线表示负数
ax.contour(x,y,z,colors='black')
```

```python
ax = plt.axes()
ax.contour(x,y,z,cmap='RdGy')
```

- 填充等高线图：`plt.contourf`。

```python
ax = plt.axes()
plt.contourf(x,y,z,cmap='RdGy')
plt.colorbar(ax=ax);
```

```python
ax = plt.axes()
# 渲染为渐变图
# extent指定[xmin,xmax,ymin,ymax]
plt.imshow(z,extent=[0,5,0,5],origin='lower',cmap='RdGy')
plt.colorbar(ax=ax)
```

- 重叠：将彩色图和等高线图叠放`plt.clabel`。

```python
fig = plt.figure()
ax = plt.axes()
contours = plt.contour(x, y, z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(z, extent=[0, 5, 0, 5], origin='lower',cmap='RdGy', alpha=0.5)
plt.colorbar();
```

`plt.contour`、`plt.contourf`、`plt.imshow`三个函数的组合可以用二维图画三维数据。

## 频次直方图、数据区间划分、分布密度

- 频次直方图：`plt.hist`。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
ax= plt.axes()
# alpha为重叠时透明度
plt.hist(data, bins=30, density=True, alpha=0.5,histtype='stepfilled', color='steelblue',edgecolor='none');
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
info = dict(alpha=0.5,histtype='stepfilled',density=True,bins=40)
plt.hist(x1,**info)
plt.hist(x2,**info)
plt.hist(x3,**info)
```

- 频次直方图计算：`np.historgram`。

```python
# bins指定个数
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
print(bin_edges)
```

- 二维频次直方图：`plt.hist2d`和六边形区间划分`plt.hexbin`。

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
# 多元高斯分布 样本数据
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
# 采用六边形区间划分
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

- 二维频次直方图计算：`np.histogram2d`。

- 核密度估计：`KDE。

```python
# 用KDE抹掉空间中离散数据点，拟合出平滑的函数
# scipy.stats中包含一个简单快速的KDE实现
from scipy.stats import gaussian_kde
# 拟合数组维度[ndim,nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)
```

```python
ax= plt.axes()
# 采用规则网格数据进行拟合
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
# KDE 方法通过不同的平滑带宽长度在拟合函数的准确性与平滑性之间作出权衡。
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))
plt.imshow(Z.reshape(Xgrid.shape),
origin='lower', aspect='auto',
extent=[-3.5, 3.5, -6, 6],
cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```

## 配置图例

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,10,1000)
fig,ax = plt.subplots()
ax.plot(x,np.sin(x),'-b',label='Sin')
ax.plot(x,np.cos(x),'--r',label='cos')
leg= ax.legend()
```

### 选择显示
`legend`默认显示所有元素的标签，如果只想显示部分，有两种解决方法：

- `plt.plot`只创建需要显示的标签。

```python
plt.plot(x,y[:,0],label='first')
plt.plot(x,y[:,1],label='second')
plt.plot(x,y[:,2:])
plt.legend(framealpha=1, frameon=True)
```

- `plt.legend`：只为需要的图例添加标签。

```python
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
# 创建了四个图
lines=plt.plot(x,y)
plt.legend(lines[:2], ['first', 'second'])
```

### 显示不同的大小的点
用不同尺寸的点表示数据的特征。

```python
import pandas as pd
cities = pd.read_csv(r"california_cities.csv")
# 提取感兴趣的数据
latitude, longtitude = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
# 画点图，c为颜色，s为面积
plt.scatter(longtitude, latitude, label=None,c=np.log10(population),s=area , cmap='viridis', linewidth=0, alpha=0.5)
plt.axis(aspect='equal')
# x轴标签
plt.xlabel('longitude')
# y轴标签
plt.ylabel('latitude')
# 创建颜色面板
plt.colorbar(label='log$_{10}$(population)')
# 设置颜色的最值
plt.clim(3, 7)
# 画三个带标签的点
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area,label=str(area) + ' km$^2$')
# 显示图例
# labelspacing表示每个图例之间的空隙
plt.legend(scatterpoints=1, frameon=True,labelspacing=1, title='City Area')
# 添加标题
plt.title('California Cities: Area and Population')
```

### 同时显示多个图例
`legend`只能为一张图创建一个图例，如果有多个图例，需要使用`ax.add_artist`添加。

```python
fig, ax = plt.subplots()
lines = []
styles = ['-', '--', '-.', ':']
colors=['r','g','b','g']
x = np.linspace(0, 10, 1000)
for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),styles[i], color=colors[i])
ax.axis('equal')
# 设置第一个图例要显示的线条和标签
ax.legend(lines[:2], ['line A', 'line B'],loc='upper right', frameon=False)
# 创建legend对象
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],loc='lower right', frameon=False)
# 添加到图中
ax.add_artist(leg);
```
